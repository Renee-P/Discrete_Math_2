**Question 1: Show that the expected value $E[y] = \Sigma y \cdot P(y) = n \cdot p_i$**


1. To find the expected value $ E[y] $, we need to compute the sum of the products of each possible outcome configuration $ y $ and its corresponding probability $ P(y) $. Using the joint probability distribution provided:

\begin{equation}
P(y) = \frac{n!}{y_1!y_2!y_3!...y_n!} \cdot p_1^{y_1} \cdot p_2^{y_2} \cdot p_3^{y_3} \cdots p_k^{y_k}
\end{equation}

2. The expected value $ E[y] $ is given by:

\begin{equation}
E[y] = \sum y \cdot P(y)
\end{equation}

3. Substituting the expression for $ P(y) $, we get:

\begin{equation}
E[y] = \sum y \cdot \frac{n!}{y_1!y_2!y_3!...y_n!} \cdot p_1^{y_1} \cdot p_2^{y_2} \cdot p_3^{y_3} \cdots p_k^{y_k}
\end{equation}

4. Since $ y_1 + y_2 + \cdots + y_k = n $, we can simplify this expression:

\begin{equation}
E[y] = \sum \left( y_1 \cdot p_1 + y_2 \cdot p_2 + \cdots + y_k \cdot p_k \right) \cdot \frac{n!}{y_1!y_2!y_3!...y_n!}
\end{equation}

5. Now, considering that $ n!/(y_1!y_2!y_3!...y_n!) $ is a constant for each $ y $, we can factor it out of the summation:

\begin{equation}
E[y] = \frac{n!}{y_1!y_2!y_3!...y_n!} \cdot \sum \left( y_1 \cdot p_1 + y_2 \cdot p_2 + \cdots + y_k \cdot p_k \right)
\end{equation}

6. Recall that $ \sum y_i = n $, therefore:

\begin{equation}
E[y] = \frac{n!}{y_1!y_2!y_3!...y_n!} \cdot \left( n \cdot p_1 + n \cdot p_2 + \cdots + n \cdot p_k \right)
\end{equation}

7. Now, $ n $ is a constant, so we can factor it out of the summation:

\begin{equation}
E[y] = \frac{n!}{y_1!y_2!y_3!...y_n!} \cdot n \cdot \left( p_1 + p_2 + \cdots + p_k \right)
\end{equation}

8. Given that $ p_1 + p_2 + \cdots + p_k = 1 $, we have:

\begin{equation}
E[y] = \frac{n!}{y_1!y_2!y_3!...y_n!} \cdot n \cdot 1
\end{equation}

9. Finally, $ n!/(y_1!y_2!y_3!...y_n!) $ can be simplified to 1 since it's a multinomial coefficient, leaving:

\begin{equation}
E[y] = n \cdot 1 = n \cdot p_i
\end{equation}

Thus, we have shown that $ E[y] = n \cdot p_i $.

**Task 2: Consider the following experiment:**

2. Each trial results in one of two outcomes: success, $S$, or failure, $F$.
3. The probability of success on a single trial is equal to some value $p$ and remains the same from trial to trial. The probability of a failure is equal to $q = (1 − p)$.
4. The trials are independent.
5. The random variable of interest is $Y$, the number of successes observed during the $n$ trials.

**Question 1: From the following steps shown above, derive the probability distribution of the experiment.**


1. To derive the probability distribution of the experiment, we'll utilize the binomial distribution since each trial results in either success or failure, and the trials are independent.

2. The probability mass function (PMF) of the binomial distribution gives the probability of observing a specific number of successes ($ Y $) in $ n $ trials, given the probability of success $ p $ and the number of trials $ n $.

The PMF of the binomial distribution is given by:

\begin{equation}
P(Y=y) = \binom{n}{y} \cdot p^y \cdot (1-p)^{n-y}
\end{equation}

where:
- $ P(Y=y) $ is the probability of observing exactly $ y $ successes,
- $ \binom{n}{y} $ is the binomial coefficient, representing the number of ways to choose $ y $ successes from $ n $ trials, calculated as $ \frac{n!}{y!(n-y)!} $,
- $ p $ is the probability of success on a single trial,
- $ (1-p) $ is the probability of failure on a single trial,
- $ y $ is the number of successes observed during the $ n $ trials.

3. Therefore, the probability distribution of the experiment is given by the binomial distribution PMF:

\begin{equation}
P(Y=y) = \binom{n}{y} \cdot p^y \cdot (1-p)^{n-y}
\end{equation}

This distribution describes the likelihood of observing each possible number of successes ($ y $) in $ n $ trials, given the probability of success $ p $ and the total number of trials $ n $.


**Question 2: Show that the expectation of this probability distribution is $E[y] = \Sigma y \cdot P(y) = np$.**


1. To show that the expectation of this probability distribution is $E[y] = \sum y \cdot P(y) = np$, we need to compute the sum of the products of each possible outcome $y$ and its corresponding probability $P(y)$, and demonstrate that it equals $np$, where $n$ is the number of trials and $p$ is the probability of success on a single trial.

2. Given the probability distribution of the experiment described by the binomial distribution PMF:

\begin{equation}
P(Y=y) = \binom{n}{y} \cdot p^y \cdot (1-p)^{n-y}
\end{equation}

3. The expectation $E[y]$ is given by:

\begin{equation}
E[y] = \sum y \cdot P(y)
\end{equation}

4. Substituting the expression for $P(Y=y)$, we get:

\begin{equation}
E[y] = \sum y \cdot \binom{n}{y} \cdot p^y \cdot (1-p)^{n-y}
\end{equation}

5. Expanding this summation over all possible values of $y$, from 0 to $n$, we have:

\begin{equation}
E[y] = \sum_{y=0}^{n} y \cdot \binom{n}{y} \cdot p^y \cdot (1-p)^{n-y}
\end{equation}

6. Now, we'll simplify this expression step by step.

- Expand the binomial coefficient $\binom{n}{y}$:

\begin{equation}
E[y] = \sum_{y=0}^{n} y \cdot \frac{n!}{y!(n-y)!} \cdot p^y \cdot (1-p)^{n-y}
\end{equation}

- Simplify the expression by canceling out $y$ in the numerator and $y!(n-y)!$ in the denominator:

\begin{equation}
E[y] = \sum_{y=0}^{n} \frac{n!}{(y-1)!(n-y)!} \cdot p^y \cdot (1-p)^{n-y}
\end{equation}

- This simplifies to:

\begin{equation}
E[y] = n \cdot \sum_{y=1}^{n} \frac{(n-1)!}{(y-1)!(n-y)!} \cdot p^y \cdot (1-p)^{n-y}
\end{equation}

- Recognize that the summation term is simply the probability distribution of a binomial distribution with $n-1$ trials, since it starts from $y=1$ and goes to $n$. Therefore, the sum is equal to 1:

\begin{equation}
E[y] = n \cdot 1 = np
\end{equation}

Thus, we have shown that $E[y] = np$, as expected. This demonstrates that the expectation of the binomial distribution is indeed $np$.


**Question 3: Experience has shown that 30% of all persons afflicted by a certain illness recover. A drug company has developed a new medication. Ten people with the illness were selected at random and received the medication; nine recovered shortly thereafter. Suppose that the medication was absolutely worthless. What is the probability that at least nine of ten receiving the medication will recover?**

1. To find the probability that at least nine of ten people receiving the medication will recover, we'll use the binomial distribution.

2. Given that 30% of all persons afflicted by the illness recover, the probability of success (recovery) on a single trial is $p = 0.30$, and the probability of failure (not recovering) is $q = 1 - p = 0.70$.

3. The probability mass function (PMF) of the binomial distribution gives the probability of observing a specific number of successes ($Y$) in ten trials, given by:

\begin{equation}
P(Y=y) = \binom{10}{y} \cdot p^y \cdot q^{10-y}
\end{equation}

where:
- $P(Y=y)$ is the probability of observing exactly $y$ successes,
- $\binom{10}{y}$ is the binomial coefficient, representing the number of ways to choose $y$ successes from ten trials,
- $p$ is the probability of success (recovery) on a single trial,
- $q$ is the probability of failure (not recovering) on a single trial,
- $y$ is the number of successes (recoveries) observed during the ten trials.

4. Now, we want to find the probability that at least nine out of ten people receiving the medication will recover. This includes cases where exactly nine or all ten people recover.

So, we need to calculate:

\begin{equation}
P(Y \geq 9) = P(Y=9) + P(Y=10)
\end{equation}

5. Substituting the values into the binomial distribution PMF:

\begin{align*}
P(Y=9) & = \binom{10}{9} \cdot 0.30^9 \cdot 0.70^{10-9} \\
& = \binom{10}{9} \cdot 0.30^9 \cdot 0.70^1 \\
& = 10 \cdot 0.30^9 \cdot 0.70 \\
& \approx 0.1210608
\end{align*}

\begin{align*}
P(Y=10) & = \binom{10}{10} \cdot 0.30^{10} \cdot 0.70^{10-10} \\
& = 0.30^{10} \cdot 0.70^0 \\
& = 0.30^{10} \\
& \approx 0.0000059049
\end{align*}

6. Therefore, the probability that at least nine out of ten people receiving the medication will recover is:

\begin{align*}
P(Y \geq 9) & = P(Y=9) + P(Y=10) \\
& \approx 0.1210608 + 0.0000059049 \\
& \approx 0.1210667
\end{align*}

This means there's approximately a 12.11% chance that at least nine out of ten people receiving the medication will recover.

**Task 3: Consider a probability distribution of $p(y) = \frac{\lambda^y}{y!} e^{-\lambda}$.**


**Question 1: Find a general formula for the expected value of this distribution.**


1. To find the expected value (mean) of the probability distribution $p(y) = \frac{\lambda^y}{y!} e^{-\lambda}$, we need to compute the sum of each possible outcome $y$ weighted by its probability $p(y)$.

2. The expected value $E[y]$ is given by:

\begin{equation}
E[y] = \sum_{y=0}^{\infty} y \cdot p(y)
\end{equation}

3. Substituting the expression for $p(y)$, we get:

\begin{equation}
E[y] = \sum_{y=0}^{\infty} y \cdot \frac{\lambda^y}{y!} e^{-\lambda}
\end{equation}

4. Now, we'll simplify this expression step by step.

- Move the constant term $\frac{\lambda^y}{y!} e^{-\lambda}$ out of the summation since it does not depend on $y$:

\begin{equation}
E[y] = e^{-\lambda} \sum_{y=0}^{\infty} y \cdot \frac{\lambda^y}{y!}
\end{equation}

- Recognize that the summation term is the series expansion of the exponential function:

\begin{equation}
e^{\lambda} = \sum_{y=0}^{\infty} \frac{\lambda^y}{y!}
\end{equation}

- Differentiate both sides of the equation with respect to $\lambda$ to find the derivative of the series expansion:

\begin{equation}
\frac{d}{d\lambda} e^{\lambda} = \frac{d}{d\lambda} \sum_{y=0}^{\infty} \frac{\lambda^y}{y!}
\end{equation}

- The derivative of $e^{\lambda}$ with respect to $\lambda$ is simply $e^{\lambda}$. On the right-hand side, the derivative and the summation can be interchanged:

\begin{equation}
e^{\lambda} = \sum_{y=0}^{\infty} \frac{d}{d\lambda} \frac{\lambda^y}{y!}
\end{equation}

- Recognize that the derivative of $\frac{\lambda^y}{y!}$ with respect to $\lambda$ is simply $\frac{y\lambda^{y-1}}{y!}$:

\begin{equation}
e^{\lambda} = \sum_{y=0}^{\infty} \frac{y\lambda^{y-1}}{y!}
\end{equation}

- Simplify the expression by canceling out $y$ in the numerator and $y!$ in the denominator:

\begin{equation}
e^{\lambda} = \sum_{y=0}^{\infty} \frac{\lambda^{y-1}}{(y-1)!}
\end{equation}

- Shift the index of summation to make it start from 1 instead of 0:

\begin{equation}
e^{\lambda} = \sum_{y=1}^{\infty} \frac{\lambda^{y-1}}{(y-1)!}
\end{equation}

- Now, we recognize that the summation term is the series expansion of the exponential function:

\begin{equation}
e^{\lambda} = e^{\lambda}
\end{equation}

- Therefore, the series expansion of $e^{\lambda}$ is indeed:

\begin{equation}
\sum_{y=1}^{\infty} \frac{\lambda^{y-1}}{(y-1)!} = e^{\lambda}
\end{equation}

5. Now, we have found that:

\begin{equation}
\sum_{y=0}^{\infty} y \cdot \frac{\lambda^y}{y!} = \lambda \cdot e^{\lambda}
\end{equation}

6. Therefore, the expected value (mean) of the probability distribution is:

\begin{equation}
E[y] = \lambda \cdot e^{-\lambda} \cdot e^{\lambda}
\end{equation}

\begin{equation}
E[y] = \lambda
\end{equation}

Thus, the general formula for the expected value of this distribution is $E[y] = \lambda$.


**Question 2: A certain type of tree has seedlings randomly dispersed in a large area, with the mean density of seedlings being approximately five per square yard. If a forester randomly locates ten 1-square-yard sampling regions in the area, find the probability that none of the regions will contain seedlings.**

1. To find the probability that none of the ten 1-square-yard sampling regions will contain seedlings, we can model this situation using a Poisson distribution.

2. Given that the mean density of seedlings is approximately five per square yard, the average number of seedlings per 1-square-yard region is $\lambda = 5$.

3. The probability mass function (PMF) of the Poisson distribution gives the probability of observing $k$ seedlings in a 1-square-yard region, given by:

\begin{equation}
P(X=k) = \frac{e^{-\lambda} \lambda^k}{k!}
\end{equation}

where:
- $P(X=k)$ is the probability of observing $k$ seedlings,
- $\lambda$ is the mean number of seedlings per 1-square-yard region,
- $k$ is the number of seedlings observed in the region.

4. Now, we want to find the probability that none of the ten regions will contain seedlings. This means each region has $k = 0$ seedlings.

So, we need to calculate:

\begin{equation}
P(X=0)^{10}
\end{equation}

5. Substituting $\lambda = 5$ and $k = 0$ into the PMF:

\begin{align*}
P(X=0) & = \frac{e^{-5} \cdot 5^0}{0!} \\
& = e^{-5}
\end{align*}

6. Therefore, the probability that none of the ten regions will contain seedlings is:

\begin{align*}
P(X=0)^{10} & = (e^{-5})^{10} \\
& = e^{-50}
\end{align*}

This means there's approximately a $e^{-50}$ chance that none of the ten regions will contain seedlings.
