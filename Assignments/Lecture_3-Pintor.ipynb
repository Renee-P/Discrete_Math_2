<h1>The Handshake Theorem</h1>

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a graph object
G = nx.Graph()

# Add nodes to the graph
G.add_nodes_from([1, 2, 3, 4])

# Define the edges of the graph
edgelist = [(1, 2), (1, 3), (2, 3), (2, 4), (3, 4)]

# Add the edges to the graph
G.add_edges_from(edgelist)

# Define the layout for visualization
pos = nx.spring_layout(G)

# Draw the graph
nx.draw(G, pos=pos, with_labels=True)

# Display the graph
plt.show()
```

To illustrate the Handshake Theorem, I generated a simple graph and verify that the sum of the degrees of all vertices equals twice the number of edges. In this example, we have a simple graph with 4 vertices and 5 edges. Now, let's verify the Handshake Theorem:

```python
# Calculate the total degree of the graph
total_degree = sum(dict(G.degree()).values())

# Calculate the number of edges in the graph
num_edges = G.number_of_edges()

# Verify the Handshake Theorem
handshake_theorem_verified = total_degree == 2 * num_edges
handshake_theorem_verified
```

**Explanation of the Handshake Theorem:**

The Handshake Theorem states that in any graph $G$, the sum of the degrees of all the vertices equals twice the number of edges of $G$. Mathematically, if the vertices of $G$ are $v_1, v_2, ..., v_n$, then the total degree of $G = \text{deg}(v_1) + \text{deg}(v_2) + \ldots + \text{deg}(v_n) = 2 \cdot$ (the number of edges of $G$).

This theorem can be used to determine if a graph exists, as the total degree should always be even.

**Properties of the Handshake Theorem:**

1. In any graph, there are an even number of vertices of odd degree.
2. The total degree of a graph is always even.

The example above returned True, confirming that the sum of the degrees of all vertices equals twice the number of edges, as stated in the Handshake Theorem.

<h1>Travelling in a Graph</h1>

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a graph object
G = nx.Graph()

# Add nodes to the graph
G.add_nodes_from([1, 2, 3])

# Define the edges of the graph
edgelist = [(1, 2), (2, 3), (1, 3)]

# Add the edges to the graph
G.add_edges_from(edgelist)

# Define the layout for visualization
pos = nx.circular_layout(G)

# Draw the graph
nx.draw(G, pos=pos, with_labels=True, node_color='lightblue', node_size=800, font_size=12, font_weight='bold')

# Add edge labels
nx.draw_networkx_edge_labels(G, pos=pos, edge_labels={(1, 2): 'e1', (2, 3): 'e2', (1, 3): 'e3'}, font_color='red')

# Display the graph
plt.title('Graph G')
plt.show()
```

This code creates a simple graph with 3 vertices and 3 edges, labeled $ e1, e2, e3$, respectively.

Now, let's illustrate each type of traveling:

1. Walk from v to w:

```python
walk_edges = [(1, 2), (2, 3)]
nx.draw(G, pos=pos, with_labels=True, node_color='lightblue', node_size=800, font_size=12, font_weight='bold')
nx.draw_networkx_edges(G, pos=pos, edgelist=walk_edges, edge_color='red', width=2)
plt.title('Walk from v1 to v3')
plt.show()
```

```python
trail_edges = [(1, 2), (2, 3)]
nx.draw(G, pos=pos, with_labels=True, node_color='lightblue', node_size=800, font_size=12, font_weight='bold')
nx.draw_networkx_edges(G, pos=pos, edgelist=trail_edges, edge_color='red', width=2)
plt.title('Trail from v1 to v3')
plt.show()
```

```python
path_edges = [(1, 2), (2, 3)]
nx.draw(G, pos=pos, with_labels=True, node_color='lightblue', node_size=800, font_size=12, font_weight='bold')
nx.draw_networkx_edges(G, pos=pos, edgelist=path_edges, edge_color='red', width=2)
plt.title('Path from v1 to v3')
plt.show()
```

```python
closed_walk_edges = [(1, 2), (2, 3), (3, 1)]
nx.draw(G, pos=pos, with_labels=True, node_color='lightblue', node_size=800, font_size=12, font_weight='bold')
nx.draw_networkx_edges(G, pos=pos, edgelist=closed_walk_edges, edge_color='red', width=2)
plt.title('Closed Walk starting and ending at v1')
plt.show()
```

<h1>Connectedness as a Graph</h1>

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a connected graph
G_connected = nx.Graph()
G_connected.add_nodes_from([1, 2, 3, 4])
G_connected.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])

# Create a disconnected graph
G_disconnected = nx.Graph()
G_disconnected.add_nodes_from([1, 2, 3, 4, 5])
G_disconnected.add_edges_from([(1, 2), (3, 4)])

# Create a graph with a circuit
G_circuit = nx.Graph()
G_circuit.add_nodes_from([1, 2, 3, 4, 5])
G_circuit.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 5), (5, 1)])

# Remove an edge from the circuit
G_circuit.remove_edge(1, 2)

# Draw the graphs
plt.figure(figsize=(15, 5))

plt.subplot(131)
nx.draw(G_connected, with_labels=True, node_color='lightblue', node_size=500)
plt.title('Connected Graph')

plt.subplot(132)
nx.draw(G_disconnected, with_labels=True, node_color='lightgreen', node_size=500)
plt.title('Disconnected Graph')

plt.subplot(133)
nx.draw(G_circuit, with_labels=True, node_color='lightcoral', node_size=500)
plt.title('Graph with Circuit (Edge Removed)')

plt.show()
```

- **Connected Graph:**
A connected graph is one where there exists a path between every pair of vertices. In the provided example, the graph has vertices 1, 2, 3, and 4, and edges connecting them in a cycle, forming a connected graph.

- **Disconnected Graph:**
A disconnected graph is one where there exist vertices that are not connected by any path. In the provided example, the graph has vertices 1, 2, 3, 4, and 5, but only vertices 1, 2, 3, and 4 are connected, leaving vertex 5 disconnected.

- **Graph with a Circuit:**
A graph with a circuit is one that contains a closed loop of edges. In the provided example, the graph has vertices 1, 2, 3, 4, and 5, with edges forming a circuit (1-2-3-4-5-1). If an edge is removed from the circuit (e.g., edge between vertices 1 and 2), the graph still remains connected, and there exists a trail between any pair of vertices.

These examples illustrate the concepts of connectedness in graphs, where the existence of paths, trails, and circuits determines the connectivity of the graph.


<h1>Connected Components</h1>

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a graph with multiple connected components
G = nx.Graph()
G.add_nodes_from([1, 2, 3, 4, 5, 6, 7, 8, 9])
G.add_edges_from([(1, 2), (2, 3), (3, 1), (4, 5), (5, 6), (6, 4), (7, 8)])

# Extract connected components
connected_components = list(nx.connected_components(G))

# Draw each connected component separately
plt.figure(figsize=(12, 6))
for i, component in enumerate(connected_components, start=1):
    plt.subplot(1, len(connected_components), i)
    subgraph = G.subgraph(component)
    nx.draw(subgraph, with_labels=True, node_color='lightblue', node_size=500)
    plt.title(f'Component {i}')

plt.tight_layout()
plt.show()
```

- **Connected Components:**
Connected components are subgraphs within a larger graph where each vertex is reachable from every other vertex in the same component. In the provided example, the graph has multiple connected components:

    - **Component 1:** Contains vertices {1, 2, 3} connected in a cycle.
    - **Component 2:** Contains vertices {4, 5, 6} connected in a cycle.
    - **Component 3:** Contains vertices {7, 8} connected by a single edge.
    - **Component 4:** Contains vertex 9, isolated from other vertices.

These connected components satisfy the criteria:
- Each component is a subgraph of the original graph.
- Each component is connected, meaning there is a path between every pair of vertices within the component.
- No other connected subgraph in the original graph contains vertices or edges that are part of any of these components.

This illustrates the concept of connected components in graphs, where each component represents a maximal set of vertices that are mutually reachable.

- **Euler's Trail in the Graph:**
  - In the provided graph, we can observe Euler's trail, which is a sequence of adjacent edges and vertices that starts at one vertex and ends at another. This trail traverses through every vertex of the graph exactly once and covers every edge exactly once. In our example, the trail starts at vertex 1, moves to vertex 2, then 3, 4, 5, and finally returns to vertex 1, forming a closed loop. This satisfies the definition of Euler's trail.

- **Euler's Path Conditions:**
  - For there to be an Euler's path from one vertex to another in a graph, specific conditions must be met:
    1. **Connectedness:** The graph must be connected, ensuring there is a path between every pair of vertices. In our example, all vertices are interconnected, forming a single connected component.
    2. **Vertex Degrees:**
       - The starting and ending vertices must have odd degrees to serve as the beginning and end of the path. In our graph, vertices 1 and 5 have odd degrees, which allows them to act as the start and end points of the path.
       - All other vertices in the graph must have positive even degrees to ensure there are enough edges to traverse while maintaining the property of passing through each vertex exactly once. In our example, vertices 2, 3, and 4 have even degrees, facilitating the traversal of edges.


<h1>Hamiltonian Circuits</h1>

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create the main graph object
G = nx.Graph()

# Add nodes to the graph
G.add_nodes_from([1, 2, 3, 4, 5, 6])

# Define the edges of the graph
edgelist = [(1, 2), (1, 3), (1, 5), (2, 3), (2, 4), (3, 4), (3, 5), (4, 6), (5, 6)]

# Add the edges to the graph
G.add_edges_from(edgelist)

# Define the layout for visualization
pos = {1: (0, 1), 2: (1, 1), 3: (0.5, 0.5), 4: (1.5, 0.5), 5: (0, 0), 6: (1, 0)}

# Draw the main graph
plt.figure(figsize=(10, 5))
plt.subplot(121)
nx.draw(G, pos=pos, with_labels=True, node_color='lightblue', node_size=1000, font_size=12, font_weight='bold')
plt.title("Main Graph")

# Add edges labels
edge_labels = {(1, 2): '1', (1, 3): '2', (1, 5): '3', (2, 3): '4', (2, 4): '5', (3, 4): '6', (3, 5): '7', (4, 6): '8', (5, 6): '9'}
nx.draw_networkx_edge_labels(G, pos=pos, edge_labels=edge_labels)

# Create a subgraph with Hamiltonian circuit properties
H = G.copy()  # Create a copy of the main graph

# Define the edges to keep in the subgraph
edges_to_keep = [(1, 2), (1, 5), (2, 3), (3, 4), (4, 6), (5, 6)]

# Remove edges not part of the Hamiltonian circuit
edges_to_remove = set(G.edges()) - set(edges_to_keep)
H.remove_edges_from(edges_to_remove)

# Draw the subgraph H
plt.subplot(122)
nx.draw(H, pos=pos, with_labels=True, node_color='lightgreen', node_size=1000, font_size=12, font_weight='bold')
plt.title("Subgraph")

# Display the main graph and the subgraph H
plt.tight_layout()
plt.show()
```

**Hamiltonian Circuit in the Graph:**

A Hamiltonian circuit in this graph would be a sequence of adjacent vertices and edges in which every vertex appears exactly once, except for the first and last, which are the same. In this example, a Hamiltonian circuit could be:

1 - 2 - 4 - 6 - 5 - 3 - 1

**Subgraph:**
We can verify if the main graph has a hamiltonian circuit if it has a subgraph with the properties:
- Contains Every Vertex: The subgraph $H$ contains every vertex of $G$, as all vertices are part of the Hamiltonian circuit.
- Connected: $H$ is connected, as the Hamiltonian circuit traverses through every vertex of the graph.
- Same Number of Edges as Vertices: Since the Hamiltonian circuit forms a cycle, the number of edges in $H$ is equal to the number of vertices.
- Every Vertex has Degree 2: In the Hamiltonian circuit, every vertex (except the first and last) is incident to exactly two edges, maintaining the degree-2 property.

And as shown in the example, it does. Thus, this subgraph satisfies the properties required for a Hamiltonian circuit.

<h1>The Travelling Salesman Problem</h1>

```python
import matplotlib.pyplot as plt
import networkx as nx

# Define the edges of the graph, including the extra edge to ensure a Hamiltonian circuit
edges = [['Home', 'Park'], 
         ['Home', 'School'],
         ['Home', 'Supermarket'],
         ['Park', 'School'],
         ['Park', 'Supermarket'],
         ['School', 'Supermarket'],
         ['Supermarket', 'Work'],
         ['Work', 'Home']]  # Extra edge to ensure a Hamiltonian circuit

# Create a graph object
G = nx.Graph()
G.add_edges_from(edges)

# Define the positions for nodes
pos = nx.spring_layout(G, seed=42)

# Draw the graph
plt.figure(figsize=(10, 6))
nx.draw(
    G, pos, edge_color='black', width=1, linewidths=1,
    node_size=500, node_color='pink', alpha=0.9,
    labels={node: node for node in G.nodes()}
)

# Add edge labels
edge_labels = {('Home', 'Park'):"5 mins", 
               ('Home', 'School'):"7 mins",
               ('Home', 'Supermarket'):"8 mins",
               ('Park', 'School'):"6 mins",
               ('Park', 'Supermarket'):"9 mins",
               ('School', 'Supermarket'):"12 mins",
               ('Supermarket', 'Work'):"15 mins",
               ('Work', 'Home'):"20 mins"}  # Extra edge label
nx.draw_networkx_edge_labels(
    G, pos, edge_labels=edge_labels,
    font_color='red'
)

# Hide axis
plt.axis('off')

# Show the graph
plt.show()
```

The graph represents different locations with edges indicating the travel time between them. To find the best route, we calculate the total time for each possible Hamiltonian circuit in the graph:

- **Route 1: Home -> Park -> School -> Supermarket -> Work -> Home**
  Total time = 5 + 6 + 7 + 12 + 15 + 20 = 65 mins

- **Route 2: Home -> School -> Park -> Supermarket -> Work -> Home**
  Total time = 7 + 6 + 5 + 9 + 15 + 20 = 62 mins

- **Route 3: Home -> Supermarket -> Park -> School -> Work -> Home**
  Total time = 8 + 9 + 6 + 12 + 15 + 20 = 70 mins

- **Route 4: Home -> Supermarket -> School -> Park -> Work -> Home**
  Total time = 8 + 12 + 6 + 5 + 15 + 20 = 66 mins

- **Route 5: Home -> Work -> School -> Park -> Supermarket -> Home**
  Total time = 20 + 7 + 6 + 9 + 15 = 57 mins

- **Route 6: Home -> Work -> Supermarket -> Park -> School -> Home**
  Total time = 20 + 15 + 9 + 6 + 12 = 62 mins

The best routes are **Route 5** (Home -> Work -> School -> Park -> Supermarket -> Home) it has the shortest total travel times among all the routes.

The traveling salesman problem aims to find the most efficient Hamiltonian circuit, minimizing the total distance traveled. While there's no known algorithm to solve it efficiently for all cases, there are algorithms that provide good approximations.
