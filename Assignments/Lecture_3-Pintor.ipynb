<h1>The Handshake Theorem</h1>

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.Graph()

G.add_nodes_from([1, 2, 3, 4])

edgelist = [(1, 2), (1, 3), (2, 3), (2, 4), (3, 4)]

G.add_edges_from(edgelist)

pos = nx.spring_layout(G)

nx.draw(G, pos=pos, with_labels=True)

plt.show()
```

To illustrate the Handshake Theorem, I generated a simple graph and verify that the sum of the degrees of all vertices equals twice the number of edges. In this example, we have a simple graph with 4 vertices and 5 edges. Now, let's verify the Handshake Theorem:

```python
total_degree = sum(dict(G.degree()).values())

num_edges = G.number_of_edges()

handshake_theorem_verified = total_degree == 2 * num_edges
handshake_theorem_verified
```

**Explanation of the Handshake Theorem:**

The Handshake Theorem states that in any graph $G$, the sum of the degrees of all the vertices equals twice the number of edges of $G$. Mathematically, if the vertices of $G$ are $v_1, v_2, ..., v_n$, then the total degree of $G = \text{deg}(v_1) + \text{deg}(v_2) + \ldots + \text{deg}(v_n) = 2 \cdot$ (the number of edges of $G$).

This theorem can be used to determine if a graph exists, as the total degree should always be even.

**Properties of the Handshake Theorem:**

1. In any graph, there are an even number of vertices of odd degree.
2. The total degree of a graph is always even.

The example above returned True, confirming that the sum of the degrees of all vertices equals twice the number of edges, as stated in the Handshake Theorem.

<h1>Travelling in a Graph</h1>

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.Graph()

G.add_nodes_from([1, 2, 3])

edgelist = [(1, 2), (2, 3), (1, 3)]

G.add_edges_from(edgelist)

pos = nx.circular_layout(G)

nx.draw(G, pos=pos, with_labels=True, node_color='lightblue', node_size=800, font_size=12, font_weight='bold')

nx.draw_networkx_edge_labels(G, pos=pos, edge_labels={(1, 2): 'e1', (2, 3): 'e2', (1, 3): 'e3'}, font_color='red')

plt.title('Graph G')
plt.show()
```

This code creates a simple graph with 3 vertices and 3 edges, labeled $ e1, e2, e3$, respectively.

Now, let's illustrate each type of traveling:

1. Walk from v to w:

```python
walk_edges = [(1, 2), (2, 3)]
nx.draw(G, pos=pos, with_labels=True, node_color='lightblue', node_size=800, font_size=12, font_weight='bold')
nx.draw_networkx_edges(G, pos=pos, edgelist=walk_edges, edge_color='red', width=2)
plt.title('Walk from v1 to v3')
plt.show()
```

Consider the graph $G$ with vertices $v_1, v_2, v_3$ and edges $e_1, e_2, e_3$. A walk from $v_1$ to $v_3$ could be $v_1 - e_1 - v_2 - e_2 - v_3$.

2. Trail from v to w:

```python
trail_edges = [(1, 2), (2, 3)]
nx.draw(G, pos=pos, with_labels=True, node_color='lightblue', node_size=800, font_size=12, font_weight='bold')
nx.draw_networkx_edges(G, pos=pos, edgelist=trail_edges, edge_color='red', width=2)
plt.title('Trail from v1 to v3')
plt.show()
```

In the same graph $G$, a trail from $v_1$ to $v_3$ could be $v_1 - e_1 - v_2 - e_3 - v_3$, where each edge is unique.

3. Path from v to w:

```python
path_edges = [(1, 2), (2, 3)]
nx.draw(G, pos=pos, with_labels=True, node_color='lightblue', node_size=800, font_size=12, font_weight='bold')
nx.draw_networkx_edges(G, pos=pos, edgelist=path_edges, edge_color='red', width=2)
plt.title('Path from v1 to v3')
plt.show()
```

Continuing with graph $G$, a path from $v_1$ to $v_3$ could be $v_1 - e_1 - v_2 - e_2 - v_3$, where each vertex is unique.

4. Closed walk:

```python
closed_walk_edges = [(1, 2), (2, 3), (3, 1)]
nx.draw(G, pos=pos, with_labels=True, node_color='lightblue', node_size=800, font_size=12, font_weight='bold')
nx.draw_networkx_edges(G, pos=pos, edgelist=closed_walk_edges, edge_color='red', width=2)
plt.title('Closed Walk starting and ending at v1')
plt.show()
```

In graph $G$, a closed walk starting and ending at $v_1$ could be $v_1 - e_1 - v_2 - e_3 - v_3 - e_2 - v_1$.

5. Circuit:

```python
circuit_edges = [(1, 2), (2, 3), (3, 1)]
nx.draw(G, pos=pos, with_labels=True, node_color='lightblue', node_size=800, font_size=12, font_weight='bold')
nx.draw_networkx_edges(G, pos=pos, edgelist=circuit_edges, edge_color='red', width=2)
plt.title('Circuit starting and ending at v1')
plt.show()
```

Using the same closed walk example, $v_1 - e_1 - v_2 - e_3 - v_3 - e_2 - v_1$, it forms a circuit since it starts and ends at the same vertex and contains at least one edge.

6. Simple circuit:

```python
simple_circuit_edges = [(1, 2), (2, 3), (3, 1)]
nx.draw(G, pos=pos, with_labels=True, node_color='lightblue', node_size=800, font_size=12, font_weight='bold')
nx.draw_networkx_edges(G, pos=pos, edgelist=simple_circuit_edges, edge_color='red', width=2)
plt.title('Simple Circuit starting and ending at v1')
plt.show()
```

<h1>Hamiltonian Circuits</h1>

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.Graph()

G.add_nodes_from([1, 2, 3, 4, 5, 6])

edgelist = [(1, 2), (1, 3), (1, 5), (2, 3), (2, 4), (3, 4), (3, 5), (4, 6), (5, 6)]

G.add_edges_from(edgelist)

pos = nx.spring_layout(G)

plt.figure(figsize=(10, 5))
plt.subplot(121)
nx.draw(G, pos=pos, with_labels=True, node_color='lightblue', node_size=1000, font_size=12, font_weight='bold')
plt.title("Main Graph G")

H = G.copy()  

edges_to_keep = [(1, 2), (1, 5), (2, 3), (3, 4), (4, 6), (5, 6)]

edges_to_remove = set(G.edges()) - set(edges_to_keep)
H.remove_edges_from(edges_to_remove)

plt.subplot(122)
nx.draw(H, pos=pos, with_labels=True, node_color='lightgreen', node_size=1000, font_size=12, font_weight='bold')
plt.title("Subgraph H")

plt.tight_layout()
plt.show()
```

**Hamiltonian Circuit in the Graph:**

A Hamiltonian circuit in this graph would be a sequence of adjacent vertices and edges in which every vertex appears exactly once, except for the first and last, which are the same. In this example, a Hamiltonian circuit could be:

1 - 2 - 3 - 4 - 6 - 5 - 1

**Subgraph:**
We can verify if the main graph has a hamiltonian circuit if it has a subgraph with the properties:
- Contains Every Vertex: The subgraph $H$ contains every vertex of $G$, as all vertices are part of the Hamiltonian circuit.
- Connected: $H$ is connected, as the Hamiltonian circuit traverses through every vertex of the graph.
- Same Number of Edges as Vertices: Since the Hamiltonian circuit forms a cycle, the number of edges in $H$ is equal to the number of vertices.
- Every Vertex has Degree 2: In the Hamiltonian circuit, every vertex (except the first and last) is incident to exactly two edges, maintaining the degree-2 property.

And as shown in the example, it does. Thus, this subgraph satisfies the properties required for a Hamiltonian circuit.

<h1>The Travelling Salesman Problem</h1>

```python
import matplotlib.pyplot as plt
import networkx as nx

edges = [['Home', 'Park'], 
         ['Home', 'School'],
         ['Home', 'Supermarket'],
         ['Park', 'School'],
         ['Park', 'Supermarket'],
         ['School', 'Supermarket'],
         ['Supermarket', 'Work'],
         ['Work', 'Home']]  

G = nx.Graph()
G.add_edges_from(edges)

pos = nx.spring_layout(G, seed=42)

plt.figure(figsize=(10, 6))
nx.draw(
    G, pos, edge_color='black', width=1, linewidths=1,
    node_size=500, node_color='pink', alpha=0.9,
    labels={node: node for node in G.nodes()}
)

edge_labels = {('Home', 'Park'):"5 mins", 
               ('Home', 'School'):"7 mins",
               ('Home', 'Supermarket'):"8 mins",
               ('Park', 'School'):"6 mins",
               ('Park', 'Supermarket'):"9 mins",
               ('School', 'Supermarket'):"12 mins",
               ('Supermarket', 'Work'):"15 mins",
               ('Work', 'Home'):"20 mins"}  
nx.draw_networkx_edge_labels(
    G, pos, edge_labels=edge_labels,
    font_color='red'
)

plt.axis('off')

plt.show()
```

The graph represents different locations with edges indicating the travel time between them. To find the best route, we calculate the total time for each possible Hamiltonian circuit in the graph:

- **Route 1: Home -> Park -> School -> Supermarket -> Work -> Home**
  Total time = 5 + 6 + 7 + 12 + 15 + 20 = 65 mins

- **Route 2: Home -> School -> Park -> Supermarket -> Work -> Home**
  Total time = 7 + 6 + 5 + 9 + 15 + 20 = 62 mins

- **Route 3: Home -> Supermarket -> Park -> School -> Work -> Home**
  Total time = 8 + 9 + 6 + 12 + 15 + 20 = 70 mins

- **Route 4: Home -> Supermarket -> School -> Park -> Work -> Home**
  Total time = 8 + 12 + 6 + 5 + 15 + 20 = 66 mins

- **Route 5: Home -> Work -> School -> Park -> Supermarket -> Home**
  Total time = 20 + 7 + 6 + 9 + 15 = 57 mins

- **Route 6: Home -> Work -> Supermarket -> Park -> School -> Home**
  Total time = 20 + 15 + 9 + 6 + 12 = 62 mins

The best routes are **Route 5** (Home -> Work -> School -> Park -> Supermarket -> Home) it has the shortest total travel times among all the routes.

The traveling salesman problem aims to find the most efficient Hamiltonian circuit, minimizing the total distance traveled. While there's no known algorithm to solve it efficiently for all cases, there are algorithms that provide good approximations.
