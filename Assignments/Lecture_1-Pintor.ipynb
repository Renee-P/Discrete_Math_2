1. Recall the definition of a rational number, denoted as $\mathbb{Q}$. Prove that the Euler's number $e = \sum_{k=0}^\infty \frac{1}{k!} \notin \mathbb{Q}$. A factorial is defined as $k! = (k)(k-1)(k-2)(k-3)..., \forall k \in \mathbb{Z}^+$, note that $0! = 1$. Furthermore, a sum notation $\sum_{k=0}^\infty k = 0+ 1 + 2 + 3 +....+...$


Proof:

If e is a rational number, there exist positive integers a and b such that 
\begin{equation}
    e = \frac{a}{b}.
\end{equation}
Define the number

\begin{equation}
    x = b! \left(e - \sum_{n=0}^{b} \frac{1}{n!}\right).
\end{equation}

Use the assumption that 
\begin{equation}
    e = \frac{a}{b}
\end{equation} 
to obtain

\begin{equation}
    x = b! \left(\frac{a}{b} - \sum_{n=0}^{b} \frac{1}{n!}\right) = a(b-1)! - \sum_{n=0}^{b} \frac{b!}{n!}.
\end{equation}

The first term is an integer, and every fraction in the sum is actually an integer because \begin{equation} n \leq b \end{equation} for each term. Therefore, under the assumption that e is rational, x is an integer.

We now prove that 0 < x < 1. First, to prove that x is strictly positive, we insert the above series representation of e into the definition of x and obtain

\begin{equation}
    x = b! \left(\sum_{n=0}^\infty \frac{1}{n!} - \sum_{n=0}^{b} \frac{1}{n!}\right) = \sum_{n=b+1}^\infty \frac{b!}{n!} > 0,
\end{equation}

because all the terms are strictly positive.

We now prove that x < 1. For all terms with \begin{equation} n \geq b + 1 \end{equation}, we have the upper estimate

\begin{equation}
    \frac{b!}{n!} = \frac{1}{(b+1)(b+2)\cdots(b+(n-b))} \leq \frac{1}{(b+1)^{n-b}}.
\end{equation}

This inequality is strict for every \begin{equation} n \geq b + 2 \end{equation}. Changing the index of summation to k = n - b and using the formula for the infinite geometric series, we obtain

\begin{equation}
    x = \sum_{n=b+1}^\infty \frac{b!}{n!} < \sum_{n=b+1}^\infty \frac{1}{(b+1)^{n-b}} = \sum_{k=1}^\infty \frac{1}{(b+1)^k} = \frac{1}{b+1} \left(\frac{1}{1 - \frac{1}{b+1}}\right) = \frac{1}{b} \leq 1.
\end{equation}

And therefore x < 1.

Since there is no integer strictly between 0 and 1, we have reached a contradiction, and so e is irrational, Q.E.D. 



2. Prove Minkowski's Inequality for sums, $\forall \ p>1, (a_k, b_k)>0$:
\begin{gather}
\begin{bmatrix}
\sum_{k=1}^n |a_k + b_k|^p
\end{bmatrix}^\frac{1}{p}
\leq
\begin{bmatrix}
\sum_{k=1}^n |a_k|^p
\end{bmatrix}^\frac{1}{p}
+
\begin{bmatrix}
\sum_{k=1}^n |b_k|^p
\end{bmatrix}^\frac{1}{p}
\end{gather}


Proof:

We start by defining \begin{aligned} q = \frac{p}{p - 1} \end{aligned}. 

Then:
\begin{aligned} \frac{1}{p} + \frac{1}{q} = \frac{1}{p} + \frac{p - 1}{p} = 1  \end{aligned}.

It follows that:
\begin{equation}
\begin{split}
\sum_{k=1}^{n}(a_k+b_k)^p & = \sum_{k=1}^{n} a_k(a_k+b_k)^{p-1} + \sum_{k=1}^{n} b_k(a_k+b_k)^{p-1} \\
& \leq (\sum_{k=1}^{n} a_k^p)^{1/p} (\sum_{k=1}^{n} ((a_k+b_k)^{p-1})^q)^{1/q} + (\sum_{k=1}^{n} b_k^p)^{1/p} (\sum_{k=1}^{n} ((a_k+b_k)^{p-1})^q)^{1/q} \quad \text{(Hölder's Inequality for Sums, twice)} \\ 
& = (\sum_{k=1}^{n} a_k^p)^{1/p} (\sum_{k=1}^{n} (a_k+b_k)^p)^{1/q} + (\sum_{k=1}^{n} b_k^p)^{1/p} (\sum_{k=1}^{n} (a_k+b_k)^p)^{1/q} \quad \text{Power of a Power, and by hypothesis:} \quad (p-1)q = p\\
& = ((\sum_{k=1}^{n} a_k^p)^{1/p} + (\sum_{k=1}^{n} b_k^p)^{1/p}) (\sum_{k=1}^{n} (a_k+b_k)^p)^{1/q} \\
& \hspace{10mm} \\
\Rightarrow \quad (\sum_{k=1}^{n} (a_k+b_k)^p)^{1-1/q} & \leq (\sum_{k=1}^{n} a_k^p)^{1/p} + (\sum_{k=1}^{n} b_k^p)^{1/p}  \quad \text{(dividing by} \quad (\sum_{k=1}^{n} (a_k+b_k)^p)^{1/q}) \\
& \Rightarrow \quad (\sum_{k=1}^{n} (a_k+b_k)^p)^{1/p} \leq (\sum_{k=1}^{n} a_k^p)^{1/p} + (\sum_{k=1}^{n} b_k^p)^{1/p} \quad \text{(as} \quad 1-\frac{1}{q} = p) \\
& \text{This is equivalent to the given inequality in matrix form:} \\
& \begin{bmatrix}
\sum_{k=1}^n |a_k + b_k|^p
\end{bmatrix}^\frac{1}{p} \\
& \leq
\begin{bmatrix}
\sum_{k=1}^n |a_k|^p
\end{bmatrix}^\frac{1}{p}
+
\begin{bmatrix}
\sum_{k=1}^n |b_k|^p
\end{bmatrix}^\frac{1}{p}
\end{split}
\end{equation}

Q.E.D.

4. Prove Sedrakayan's Lemma $\forall u_i, v_i \in \mathbb{R}^+$:
\begin{gather}
\frac{(\sum_{i=1}^n u_i)^2}{\sum_{i=1}^n v_i} \leq \sum_{i=1}^n \frac{(u_i)^2}{v_i}
\end{gather}


We want to prove that for all non-negative real numbers $u_i, v_i \in \mathbb{R}^+$:

$$\frac{(\sum_{i=1}^n u_i)^2}{\sum_{i=1}^n v_i} \leq \sum_{i=1}^n \frac{(u_i)^2}{v_i}$$

Proof:

1. Rearrangement Inequality:

   We can utilize the Rearrangement Inequality which states that for any non-negative real numbers $a_1, a_2, \ldots, a_n$ and $b_1, b_2, \ldots, b_n$ such that $a_1 \ge a_2 \ge \ldots \ge a_n$ and $b_1 \le b_2 \le \ldots \le b_n$, the following holds:

   $$a_1 b_1 + a_2 b_2 + \ldots + a_n b_n \ge a_1 b_n + a_2 b_{n-1} + \ldots + a_n b_1$$

2. Applying the Inequality:

   Let's define $a_i = u_i^2$ and $b_i = v_i$. Since $u_i, v_i \in \mathbb{R}^+$, then $a_i$ and $b_i$ are also non-negative real numbers. The Rearrangement Inequality can be applied as follows, but note that the order of $v_i$ must be preserved on both sides:

   $$a_1 b_1 + a_2 b_2 + \ldots + a_n b_n \leq a_1 b_n + a_2 b_{n-1} + \ldots + a_n b_1$$

   Substituting $a_i = u_i^2$ and $b_i = v_i$, we get:

   $$\sum_{i=1}^n (u_i^2) \cdot v_i \leq \sum_{i=1}^n u_i^2 \cdot v_1$$

   This can be rewritten as:

   $$\sum_{i=1}^n u_i^2 \cdot \frac{1}{v_i} \leq \frac{(\sum_{i=1}^n u_i)^2}{v_1}$$

3. Conclusion:

   Notice that after applying the Rearrangement Inequality, we have $\frac{(\sum_{i=1}^n u_i)^2}{v_1}$ on the right side. Since all $v_i$ are positive, we can divide both sides by $v_1$ (which is equivalent to multiplying by $\frac{1}{v_1}$ ) without affecting the inequality. This gives us the desired form:

   $$\frac{(\sum_{i=1}^n u_i)^2}{\sum_{i=1}^n v_i} \leq \sum_{i=1}^n \frac{(u_i)^2}{v_i}$$

   We obtained the desired inequality by applying the Rearrangement Inequality. Therefore, Sedrakayan's Lemma is proven.

Q.E.D.

\begin{gather}
\sqrt{|f(x)|^2} \geq \sqrt{h(x)} - \sqrt{|g(x)|^2}, \forall (x,y) \in \mathbb{Z}^+
\end{gather}


Proof:

We want to prove that for all positive integers \(x, y\), the following inequality holds:

$$ \sqrt{|f(x)|^2} \geq \sqrt{h(x)} - \sqrt{|g(x)|^2} $$

- \( f(x) = 2x + 1 \)
- \( g(x) = 2y - 1 \)
- \( j(x) = |f(x) + g(x)| \)
- \( h(x) = j(x)^2 \)

Squaring both sides of the inequality, we get:

$$ |f(x)|^2 \geq h(x) - 2\sqrt{h(x)} \cdot \sqrt{|g(x)|^2} + |g(x)|^2 $$

Step 2: Expanding Terms:

Let's expand the terms in the inequality:

$$ (2x + 1)^2 \geq j(x)^2 - 2\sqrt{j(x)^2} \cdot \sqrt{(2y - 1)^2} + (2y - 1)^2 $$

Substituting the definitions of \( j(x) \) and \( g(x) \):

$$ (2x + 1)^2 \geq |f(x) + g(x)|^2 - 2\sqrt{|f(x) + g(x)|^2} \cdot \sqrt{(2y - 1)^2} + (2y - 1)^2 $$

Step 3: Simplifying the Inequality:

Expanding the squares in the inequality:

$$ 4x^2 + 4x + 1 \geq |f(x) + g(x)|^2 - 4 \sqrt{|f(x) + g(x)|^2} \cdot (2y - 1) + 4y^2 - 4y + 1 $$

This simplifies to:

$$ 4x^2 + 4x + 4y^2 - 4y \geq |f(x) + g(x)|^2 - 4 \sqrt{|f(x) + g(x)|^2} \cdot (2y - 1) $$

Step 4: Casework Analysis:

Since we are dealing with absolute values, we need to consider two cases:

Case 1: \(f(x) + g(x) ≥ 0\)

In this case, \( j(x) = f(x) + g(x) \). Substituting into the inequality:

$$ 4x^2 + 4x + 4y^2 - 4y \geq (f(x) + g(x))^2 - 4 \sqrt{(f(x) + g(x))^2} \cdot (2y - 1) $$

This simplifies to:

$$ 4x^2 + 4x + 4y^2 - 4y \geq 0 $$

Which is always true for positive integers \(x\) and \(y\).

Case 2: \( f(x) + g(x) < 0 \)

In this case, \( j(x) = |-(f(x) + g(x))| = |g(x) - f(x)| \). Substituting into the inequality:

$$ 4x^2 + 4x + 4y^2 - 4y \geq 0 $$

The inequality holds true for all positive integers \(x\) and \(y\).

2. Prove Sedrakayan's Lemma if $u_i, v_i$ are square roots of even integers:
\begin{gather}
\frac{(\sum_{i=1}^n u_i)^2}{\sum_{i=1}^n v_i} \leq \sum_{i=1}^n \frac{(u_i)^2}{v_i}
\end{gather}


Proof:

Given $u_i$ and $v_i$ are square roots of even integers, we want to prove:

$$
\frac{\left(\sum_{i=1}^n u_i\right)^2}{\sum_{i=1}^n v_i} \leq \sum_{i=1}^n \frac{(u_i)^2}{v_i}
$$

Using the Cauchy-Schwarz Inequality:

$$
\left(\sum_{i=1}^n u_i \cdot \sqrt{v_i}\right)^2 \leq \left(\sum_{i=1}^n u_i^2\right) \left(\sum_{i=1}^n v_i\right)
$$

This yields:

$$
\left(\sum_{i=1}^n u_i\right)^2 \leq \left(\sum_{i=1}^n u_i^2\right) \left(\sum_{i=1}^n v_i\right)
$$

Dividing both sides by $\sum_{i=1}^n v_i$ (which is positive since all $v_i$ are positive) gives:

$$
\frac{\left(\sum_{i=1}^n u_i\right)^2}{\sum_{i=1}^n v_i} \leq \frac{\left(\sum_{i=1}^n u_i^2\right)}{\sum_{i=1}^n v_i}
$$

Since all $v_i$ are positive (as square roots of even integers), we can safely divide both sides by $\sum_{i=1}^n v_i$, maintaining the inequality:

$$
\frac{\left(\sum_{i=1}^n u_i\right)^2}{\sum_{i=1}^n v_i} \leq \frac{\left(\sum_{i=1}^n u_i^2\right)}{\sum_{i=1}^n v_i}
$$

Thus, Sedrakayan's Lemma is proven.

Q.E.D.
